
Aroma-based Winery Search Find wineries producing wines with aromas of plum, cherry, rose, or hazelnut (singular form only). To make things simpler, exclude any wine descriptions that contain the plural forms (ex. cherries). Table winemag_p1




id: bigint country: string description: string designation: string points: bigint price: double province: string region_1: string region_2: string variety: string winery: string

```python
from pyspark.sql.types import StructType, StructField, LongType, StringType, DoubleType
from pyspark.sql.functions import col

data = [
    (
        126576,
        "US",
        "Rich wine with plum aroma and oak finish",
        "Estate Club",
        87,
        32.0,
        "Virginia",
        "Virginia",
        "",
        "Merlot",
        "Veramar"
    ),
    (
        127077,
        "Italy",
        "Fresh cherry aroma with creamy texture",
        "Extra Dry",
        85,
        19.0,
        "Veneto",
        "Prosecco",
        "",
        "Prosecco",
        "Varaschin"
    ),
    (
        143029,
        "US",
        "Loaded with cherries and chocolate flavors",
        "Signature",
        83,
        45.0,
        "California",
        "Paso Robles",
        "Central Coast",
        "Cabernet Sauvignon",
        "Byington"
    ),
    (
        98813,
        "US",
        "Elegant hazelnut aroma with smooth texture",
        "Reserve",
        93,
        100.0,
        "Washington",
        "Washington",
        "Washington Other",
        "Cabernet Sauvignon",
        "Boudreaux Cellars"
    ),
    (
        43172,
        "US",
        "Bright roses and citrus on the palate",
        "Les Pionniers",
        86,
        40.0,
        "California",
        "Sonoma County",
        "Sonoma",
        "Sauvignon Blanc",
        "Capture"
    )
]

schema = StructType([
    StructField("id", LongType(), True),
    StructField("country", StringType(), True),
    StructField("description", StringType(), True),
    StructField("designation", StringType(), True),
    StructField("points", LongType(), True),
    StructField("price", DoubleType(), True),
    StructField("province", StringType(), True),
    StructField("region_1", StringType(), True),
    StructField("region_2", StringType(), True),
    StructField("variety", StringType(), True),
    StructField("winery", StringType(), True)
])


df = spark.createDataFrame(data, schema)

#rlike() to apply the regex pattern to the 'description' column
regex_pattern = "(?i)\\b(plum|cherry|rose|hazelnut)\\b"
filtered_df = df.filter(col("description").rlike(regex_pattern))

# 4. Display the wineries and descriptions that match the criteria
filtered_df.select("winery", "description").show(truncate=False)
```
