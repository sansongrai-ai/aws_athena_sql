```python
import uuid
def foreach_batch_function(df, epoch_id):
    array_id = df.collect()[0][0]
    if array_id == 3:
        raise Exception("Stopping the streaming")

    df.write.mode("append").json(output_path + "/" +str(uuid.uuid4()))


query = (df.writeStream.foreachBatch(foreach_batch_function).option("checkpointLocation", checkpoint_path).start())
```
